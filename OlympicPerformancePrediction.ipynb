```python
# REFERENCE CODE FOR SYNOPSIS
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 1. DATASET CREATION
data = {
    "Country": [
        "USA", "China", "India", "UK", "Japan", "Germany",
        "Russia", "Australia", "France", "Italy",
        "USA", "China", "India", "UK", "Japan", "Germany",
        "Russia", "Australia", "France", "Italy"
    ],
    
    "Year": [
        2012, 2012, 2012, 2012, 2012, 2012,
        2012, 2012, 2012, 2012,
        2016, 2016, 2016, 2016, 2016, 2016,
        2016, 2016, 2016, 2016
    ],
    
    "Athletes": [
        530, 396, 83, 541, 293, 392,
        436, 410, 330, 290,
        554, 412, 120, 547, 338, 425,
        420, 421, 350, 310
    ],
    
    "Gold": [
        46, 38, 0, 29, 7, 11,
        24, 8, 11, 8,
        46, 26, 1, 27, 12, 17,
        19, 8, 10, 8
    ],
    
    "Silver": [
        28, 27, 2, 17, 14, 19,
        26, 15, 11, 9,
        37, 18, 2, 23, 8, 10,
        18, 11, 18, 12
    ],
    
    "Bronze": [
        29, 23, 4, 19, 17, 14,
        32, 12, 12, 11,
        38, 26, 2, 17, 21, 15,
        19, 10, 14, 8
    ]
}

df = pd.DataFrame(data)
print(df)
df

# 2. DATA PREPROCESSING

# Add Total Medals
df["Total"] = df["Gold"] + df["Silver"] + df["Bronze"]

# Handling Missing Values (example)
df.fillna(0)

# Remove duplicates
df.drop_duplicates()

# 3. NORMALIZATION (Min-Max Scaler)

df["Athletes_normalisation"] = (df["Athletes"] - df["Athletes"].min()) / (df["Athletes"].max() - df["Athletes"].min())

# 4. EXPLORATORY DATA ANALYSIS (EDA)

print("\n___Basic Info___")
print(df.info())

print("\n___Statistical Summary___")
print(df.describe())

# Group by country
country_medals = df.groupby("Country")["Total"].sum()
print("\n___Total Medals by Country___")
print(country_medals)

# 5. VISUALIZATION

# Bar Chart - Total medals by country
country_medals.plot(kind="bar")
plt.title("Total Medals by Country")
plt.xlabel("Country")
plt.ylabel("Medals")
plt.show()

# Line Plot - Performance over years (USA example)
usa_data = df[df["Country"] == "USA"]
plt.plot(usa_data["Year"], usa_data["Total"], marker='*')
plt.title("USA Performance Over Years")
plt.xlabel("Year")
plt.ylabel("Total Medals")
plt.show()

# Scatter Plot - Athletes vs Medals
plt.scatter(df["Athletes"], df["Total"])
plt.title("Athletes vs Total Medals")
plt.xlabel("Athletes")
plt.ylabel("Total Medals")
plt.show()
```
